```python
import matplotlib.pyplot as plt
from nmr_to_df import nmr_to_pandas
from rich import print
```

```python
def calc_contour_levels(contour_start=30000, contour_num=20, contour_factor=1.2):
    # calculate contour levels
    cl = contour_start * contour_factor ** np.arange(contour_num)
    return cl

def calc_threshold(intensities: pd.Series, nstd: float=5) -> float:
    threshold = intensities.median() + intensities.std() * nstd
    return threshold
```

```python
data = nmr_to_pandas("test_pipe.ft2",pseudo_dim=0)
# calculate picking and plotting threshold
threshold = calc_threshold(data.df.groupby("Z").get_group(0).intensity)
print(f"[green]Threshold[/green] = {threshold:.3e}")
# detect all peaks with a threshold
# peaks = ng.peakpick.pick(data.data[0], pthres=threshold, algorithm="thres",msep=[1,1])
```

```python
# get first plane and slice data
d = data.df.query(f"Z==0 & X_PPM > 6 & X_PPM < 10 & Y_PPM > 100 & Y_PPM < 135")
# get limits for ppm scale
X_min, X_max = d.X_PPM.min(), d.X_PPM.max()
Y_min, Y_max = d.Y_PPM.min(), d.Y_PPM.max()
# create figure
fig = plt.figure()
ax = fig.add_subplot(111)
plt.contour(d.intensity.values.reshape(len(d.Y_PPM.unique()),len(d.X_PPM.unique())),
            calc_contour_levels(contour_start=threshold),
            extent=(X_max,X_min,Y_max,Y_min))
ax.set_xlabel(data.udic[data.uc_dic["X"]["dim"]]["label"])
ax.set_ylabel(data.udic[data.uc_dic["Y"]["dim"]]["label"])
ax.invert_xaxis()
ax.invert_yaxis()
```
